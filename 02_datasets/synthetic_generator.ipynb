```python
# Import required libraries
import pandas as pd
import random
import numpy as np
```

```python
def create_test_dataset(num_rows=1000):
    """
    Create a test dataset with randomized data for testing queries
    without using PII data.
    """
    
    data = []
    
    for _ in range(num_rows):
        row = {
            # Tax year: random between 2015-2025
            'tax_year': random.randint(2015, 2025),
            
            # Zipcode: random 5-digit zipcode
            'zipcode': random.randint(10000, 99999),
            
            # Balance: random between 10k-100k
            'balance': random.randint(10000, 100000),
            
            # Add more columns as needed for your testing
            'account_type': random.choice(['checking', 'savings', 'investment']),
            'status': random.choice(['active', 'inactive', 'pending']),
            'region': random.choice(['north', 'south', 'east', 'west'])
        }
        data.append(row)
    
    # Create DataFrame
    df = pd.DataFrame(data)
    return df
```

```python
# Generate the test dataset
test_df = create_test_dataset(1000)

print("Test Dataset Created!")
print(f"Shape: {test_df.shape}")
```

```python
# Example Filter Tests

# Filter by tax year
recent_years = test_df[test_df['tax_year'] >= 2020]
print(f"Records from 2020+: {len(recent_years)}")

# Filter by balance threshold
high_balance = test_df[test_df['balance'] > 50000]
print(f"High balance accounts (>50k): {len(high_balance)}")

# Complex filter
complex_filter = test_df[
    (test_df['tax_year'] >= 2020) & 
    (test_df['balance'] > 25000) & 
    (test_df['status'] == 'active')
]
print(f"Complex filter results: {len(complex_filter)}")
```
