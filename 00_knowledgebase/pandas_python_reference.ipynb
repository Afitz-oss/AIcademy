.reset_index()
What It Does
Purpose: Resets the index of a DataFrame to the default integer index.
How It Works: By default, the current index is added as a new column. If you prefer not to keep the old index, you can pass drop=True.
When to Use It
After operations that change or set a custom index (like grouping or filtering) when you want to revert back to a simple integer index.
When you need to convert the index into a column for further analysis.

```python
import pandas as pd

# Create a sample DataFrame with a custom index
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Score': [85, 90, 95]}
df = pd.DataFrame(data)
df.index = ['a', 'b', 'c']  # Setting a custom index
print("Original DataFrame with custom index:")
print(df)

# Reset the index and keep the old index as a column
df_reset = df.reset_index()
print("\nDataFrame after reset_index() (keeping old index):")
print(df_reset)

# Reset the index and drop the old index
df_reset_drop = df.reset_index(drop=True)
print("\nDataFrame after reset_index(drop=True) (dropping old index):")
print(df_reset_drop)
```

.idxmax()
What It Does
Purpose: Returns the index label of the first occurrence of the maximum value.
For DataFrames: When used on a DataFrame, it returns a Series where each element is the index label of the maximum value in that column.
When to Use It
To quickly identify the position (index label) of the maximum value in a Series.
To determine which row (by label) has the highest value in each column of a DataFrame.

```python
# Example Series
s = pd.Series([10, 20, 30, 20, 10], index=['a', 'b', 'c', 'd', 'e'])
print("Series:")
print(s)

# Get the index of the maximum value
max_index = s.idxmax()
print("\nIndex of maximum value:", max_index)
```

```python
# Example DataFrame
df = pd.DataFrame({
    'Math': [88, 92, 79],
    'Science': [93, 88, 94]
}, index=['Alice', 'Bob', 'Charlie'])
print("\nDataFrame:")
print(df)

# Get the index labels corresponding to the maximum value in each column
max_indices = df.idxmax()
print("\nIndex labels with maximum values for each column:")
print(max_indices)
```

.idxmin()
What It Does
Purpose: Returns the index label of the first occurrence of the minimum value.
For DataFrames: When applied to a DataFrame, it returns a Series with the index labels for the minimum value in each column.
When to Use It
To quickly find the location (index label) of the smallest value in a Series.
To identify which row has the minimum value for each column in a DataFrame.

```python
# Example Series
s = pd.Series([10, 20, 5, 20, 10], index=['a', 'b', 'c', 'd', 'e'])
print("Series:")
print(s)

# Get the index of the minimum value
min_index = s.idxmin()
print("\nIndex of minimum value:", min_index)
```

```python
# Example DataFrame
df = pd.DataFrame({
    'Math': [88, 92, 79],
    'Science': [93, 88, 94]
}, index=['Alice', 'Bob', 'Charlie'])
print("\nDataFrame:")
print(df)

# Get the index labels corresponding to the minimum value in each column
min_indices = df.idxmin()
print("\nIndex labels with minimum values for each column:")
print(min_indices)
```

.nlargest()
What It Does
Purpose: Returns the n largest values from a Series (or rows from a DataFrame based on a column), sorted in descending order.
When to Use It
To extract the top “n” values (e.g., top performers, highest scores) quickly from a Series.
For DataFrames, to select the rows with the largest values in a specific column.

```python
# Example Series
s = pd.Series([15, 30, 10, 50, 40], index=['a', 'b', 'c', 'd', 'e'])
print("Series:")
print(s)

# Get the top 3 largest values
top3 = s.nlargest(3)
print("\nTop 3 largest values:")
print(top3)
```

```python
# Example DataFrame: Find rows with the top 2 scores
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Score': [85, 90, 95, 80]
})
print("\nDataFrame:")
print(df)

# Get the rows with the top 2 scores based on the 'Score' column
top2 = df.nlargest(2, 'Score')
print("\nRows with top 2 scores:")
print(top2)
```

.nsmallest()
What It Does
Purpose: Returns the n smallest values from a Series (or rows from a DataFrame based on a column), sorted in ascending order.
When to Use It
To extract the bottom “n” values (e.g., lowest scores, poorest performance) quickly from a Series.
For DataFrames, to select the rows with the smallest values in a specific column.

```python
# Example Series
s = pd.Series([15, 30, 10, 50, 40], index=['a', 'b', 'c', 'd', 'e'])
print("Series:")
print(s)

# Get the bottom 2 smallest values
bottom2 = s.nsmallest(2)
print("\nBottom 2 smallest values:")
print(bottom2)
```

```python
# Example DataFrame: Find rows with the bottom 2 scores
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Score': [85, 90, 95, 80]
})
print("\nDataFrame:")
print(df)

# Get the rows with the bottom 2 scores based on the 'Score' column
bottom2 = df.nsmallest(2, 'Score')
print("\nRows with bottom 2 scores:")
print(bottom2)
```

Summary
.reset_index(): Resets DataFrame indices; useful for reformatting your DataFrame after certain operations.
.idxmax() / .idxmin(): Locate the index label where the maximum or minimum value occurs, respectively.
.nlargest() / .nsmallest(): Quickly extract the top or bottom n values from a Series or DataFrame column.

### 2.3 Control Flow
Control flow statements help you make decisions and iterate over data.

Conditional Statements (if / elif / else):

```python
x = 10
if x > 10:
    print("x is greater than 10")
elif x == 10:
    print("x is equal to 10")
else:
    print("x is less than 10")
```

Output: "x is equal to 10"

For Loop (for i in range(n)):

```python
for i in range(5):
    print(i)
```

Output: 0, 1, 2, 3, 4 (each on a new line)

While Loop (while condition):

```python
count = 0
while count < 3:
    print(count)
    count += 1
```

```python
numbers = [1, 2, 3, 4]
squares = list(map(lambda x: x**2, numbers))
print(squares)  # Output: [1, 4, 9, 16]
```

2.10 Regular Expressions
Work with pattern matching using the re module.

Importing and Basic Matching:

```python
import re
pattern = r'\d+'
text = "There are 24 hours in a day."
match = re.search(pattern, text)
if match:
    print(match.group())  # Output: '24'
```

Finding All Occurrences:

```python
numbers = re.findall(r'\d+', "I have 2 apples and 10 oranges")
print(numbers)  # Output: ['2', '10']
```

Substitution:

```python
result = re.sub(r'apples', 'bananas', "I have 2 apples")
print(result)  # Output: "I have 2 bananas"
```

## Generator Expression 

What is a Generator Expression?
A generator expression is a concise way to create an iterator without storing all values in memory at once. It is similar to a list comprehension but more memory-efficient because it generates values on demand rather than storing them in a list.



Syntax:

This returns a generator object, which produces values one at a time when iterated over.

How is a Generator Expression Different from a List Comprehension?
Feature	Generator Expression	List Comprehension
Memory Usage	Uses less memory, generates values one by one	Stores all values in a list
Execution	Lazily evaluated (only when needed)	Evaluates all at once
Performance	Faster for large data sets	Can be slower for large data sets

```python
gen_exp = (x * 2 for x in range(5))  # Creates a generator object

print(next(gen_exp))  # Output: 0
print(next(gen_exp))  # Output: 2
print(next(gen_exp))  # Output: 4
```

```python
def square_numbers(n):
    return (x ** 2 for x in range(n))  # Generator expression

squares = square_numbers(5)

for num in squares:
    print(num)
```

```python
total = sum(x ** 2 for x in range(5))  # Efficiently calculates sum of squares
print(total)  # Output: 30
```
