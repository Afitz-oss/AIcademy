## Stage 0 — Mental Models <a id='stage-0'></a>

### 0.1 What is a computer actually doing?

A computer has three main parts that matter for programming:
- **CPU** — the brain. It executes instructions one at a time, very fast.
- **RAM** — short-term memory. It holds the data your program is currently using. When you turn off the computer, RAM is wiped.
- **Disk** — long-term memory. Files, databases — anything that persists.

When you run a Python program, the CPU reads your instructions top to bottom and executes them. Variables live in RAM while the program is running.

### 0.2 What is a variable?

A variable is a **name that points to a value stored in RAM**.

The analogy: imagine RAM as a giant whiteboard with millions of labeled boxes. When you write `x = 5`, Python finds an empty box, writes `5` in it, and sticks a label `x` on it. Whenever you use `x` later, Python looks up the box labeled `x` and reads what's inside.

```python
# Creating variables
x = 5          # x points to the integer 5
name = "Alex"  # name points to the string "Alex"
is_ready = True  # is_ready points to the boolean True

print(x)
print(name)
print(is_ready)
```

### 0.3 What is a function?

A function is a **named, reusable block of instructions**.

The analogy: a recipe. "Scrambled eggs" is a name for a specific set of steps. You can follow that recipe any time by just saying its name. You don't rewrite the steps every time.

In Python, you define the recipe once with `def`, then call it by name.

```python
# Define the recipe once
def greet(name):
    print(f"Hello, {name}!")

# Call it as many times as you want
greet("Alex")
greet("Jordan")
greet("Sam")
```

### 0.4 How to read an error message

Error messages are not failures — they are the computer telling you **exactly** what went wrong. Learning to read them is the single most important skill for beginners.

Every Python error has the same structure:
1. **File and line number** — where the error happened
2. **Arrow (^)** — pointing at the specific problem
3. **Error type** — what kind of error
4. **Error message** — what the computer expected vs. what it got

**Common error types:**

| Error | What it means | Common cause |
|---|---|---|
| `SyntaxError` | Python can't parse your code | Missing colon, mismatched parentheses |
| `NameError` | You used a name that doesn't exist | Typo in variable name, forgot to define it |
| `TypeError` | Wrong type for an operation | Trying to add a string and an integer |
| `IndexError` | List index out of range | `my_list[5]` but list only has 3 items |
| `KeyError` | Dictionary key doesn't exist | `my_dict["age"]` but no "age" key |
| `AttributeError` | Object doesn't have that method/attribute | Typo in method name |
| `IndentationError` | Indentation is wrong | Mixed tabs and spaces, missing indent |

## Stage 1 — Core Building Blocks <a id='stage-1'></a>

### 1.1 Data Types

Python has four fundamental data types every beginner must know:

```python
# int — whole numbers (no decimal)
age = 25
count = -3

# float — numbers with a decimal point
price = 9.99
temperature = -12.5

# str — text, always in quotes
name = "Alex"
message = 'Hello, world!'

# bool — True or False (capital T and F)
is_logged_in = True
has_errors = False

# Check the type of anything with type()
print(type(age))          # <class 'int'>
print(type(price))        # <class 'float'>
print(type(name))         # <class 'str'>
print(type(is_logged_in)) # <class 'bool'>
```

```python
# IMPORTANT: "5" and 5 are completely different things
num = 5
text = "5"

print(num + 10)   # 15 — math works
# print(text + 10)  # TypeError! Can't add str and int

# Converting between types
print(int("5") + 10)   # 15 — convert string to int first
print(str(5) + " items")  # "5 items" — convert int to string
```

### 1.2 Arithmetic and Comparison Operators

```python
# Arithmetic operators
print(10 + 3)   # 13  — addition
print(10 - 3)   # 7   — subtraction
print(10 * 3)   # 30  — multiplication
print(10 / 3)   # 3.333... — division (always returns float)
print(10 // 3)  # 3   — floor division (integer result, rounds down)
print(10 % 3)   # 1   — modulo (remainder after division)
print(10 ** 3)  # 1000 — exponentiation (10 to the power of 3)
```

```python
# Comparison operators — always return True or False
x = 10
print(x == 10)  # True  — equal to
print(x != 10)  # False — not equal to
print(x > 5)    # True  — greater than
print(x < 5)    # False — less than
print(x >= 10)  # True  — greater than or equal
print(x <= 9)   # False — less than or equal

# CRITICAL: = is assignment, == is comparison
x = 5    # "put 5 in the box labeled x"
x == 5   # "is the value in x equal to 5?" → True
```

### 1.3 Conditional Logic

```python
# Basic if/elif/else
score = 85

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else:
    print("Below C")
```

```python
# Combining conditions with and, or, not
age = 22
has_id = True

if age >= 21 and has_id:
    print("Access granted")

temperature = 95
if temperature > 90 or temperature < 32:
    print("Extreme weather")

is_raining = False
if not is_raining:
    print("Good day for a walk")
```

## Stage 2 — Collections and Repetition <a id='stage-2'></a>

### 2.1 Lists

```python
# Creating a list
fruits = ["apple", "banana", "cherry"]
numbers = [1, 2, 3, 4, 5]
mixed = [1, "hello", True, 3.14]  # lists can hold any type

# Indexing — starts at 0, not 1
print(fruits[0])   # "apple"  (first item)
print(fruits[1])   # "banana" (second item)
print(fruits[-1])  # "cherry" (last item — negative indexing)

# Slicing — [start:stop] — stop is exclusive
print(fruits[0:2])  # ['apple', 'banana'] — items 0 and 1, not 2
print(fruits[1:])   # ['banana', 'cherry'] — from index 1 to end
print(fruits[:2])   # ['apple', 'banana'] — from start to index 1
```

```python
# Mutating a list
fruits = ["apple", "banana", "cherry"]

fruits.append("date")       # add to end
fruits.insert(1, "avocado") # insert at index 1
fruits.remove("banana")     # remove by value
last = fruits.pop()         # remove and return last item
fruits.sort()               # sort in place (alphabetically)

print(len(fruits))          # number of items
print("apple" in fruits)    # True — membership test
```

### 2.2 Dictionaries

```python
# Creating a dictionary
person = {
    "name": "Alex",
    "age": 25,
    "city": "New York"
}

# Accessing values
print(person["name"])          # "Alex" — raises KeyError if key doesn't exist
print(person.get("age"))       # 25    — returns None if key doesn't exist (safe)
print(person.get("job", "N/A")) # "N/A" — default value if key missing

# Adding and updating
person["job"] = "Engineer"    # add new key
person["age"] = 26            # update existing key

# Iterating
for key in person.keys():
    print(key)

for value in person.values():
    print(value)

for key, value in person.items():
    print(f"{key}: {value}")
```

### 2.3 for Loops

```python
# Loop over a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

# Loop with index using enumerate()
for index, fruit in enumerate(fruits):
    print(f"{index}: {fruit}")

# Loop over a range
for i in range(5):       # 0, 1, 2, 3, 4
    print(i)

for i in range(1, 6):    # 1, 2, 3, 4, 5
    print(i)

for i in range(0, 10, 2): # 0, 2, 4, 6, 8 (step of 2)
    print(i)
```

### 2.4 while Loops

```python
# while loop — runs as long as the condition is True
count = 0
while count < 5:
    print(count)
    count += 1  # ALWAYS update the condition variable — or it loops forever!

# break — exit the loop immediately
count = 0
while True:
    if count >= 5:
        break
    print(count)
    count += 1

# continue — skip to the next iteration
for i in range(10):
    if i % 2 == 0:  # skip even numbers
        continue
    print(i)  # only prints odd numbers
```

## Stage 3 — Functions and Reusability <a id='stage-3'></a>

### 3.1 Defining and Calling Functions

```python
# Define a function
def add_numbers(a, b):
    result = a + b
    return result

# Call it
total = add_numbers(3, 4)
print(total)  # 7

# Parameters: a, b (placeholders in the definition)
# Arguments: 3, 4 (actual values when calling)
```

### 3.2 return vs. print() — The Most Important Distinction

This is the #1 thing that confuses beginners.

```python
# print() — shows something to YOU, the human. The function gives nothing back.
def bad_add(a, b):
    print(a + b)  # you see the result, but it disappears

result = bad_add(3, 4)  # prints 7
print(result)           # None — because the function returned nothing!

# ---

# return — sends the value BACK to whoever called the function
def good_add(a, b):
    return a + b  # hands the result back

result = good_add(3, 4)  # result now holds 7
print(result)            # 7 — we can use it!
doubled = result * 2     # 14 — we can do more with it!
```

### 3.3 Scope

```python
# Variables inside a function only exist inside that function
def my_function():
    inside_var = "I'm inside"
    print(inside_var)  # works fine

my_function()
# print(inside_var)  # NameError! inside_var doesn't exist out here

# Why this is a FEATURE: functions are self-contained.
# You can use the same variable names in different functions without conflict.
```

### 3.4 Default Parameters

```python
# Default parameters — used when the caller doesn't provide that argument
def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}!")

greet("Alex")               # Hello, Alex!
greet("Jordan", "Hey")      # Hey, Jordan!
greet("Sam", greeting="Hi") # Hi, Sam! (keyword argument)

# Rule: default parameters must come AFTER non-default ones
# def bad(greeting="Hello", name):  # SyntaxError!
# def good(name, greeting="Hello"): # Correct
```
